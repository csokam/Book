# Telex főoldalán található hírek kigyűjtése

A <a href="www.telex.hu">telex.hu</a> struktúrája viszonylag átlátható, gyakorlásra megfelelő alapot biztosít. 

## Szükséges könyvtárak letöltése, importálása

```python
# %pip install bs4
```

```python
import os
import csv
import requests
from bs4 import BeautifulSoup as bs
```

## Weboldal HTML struktúrájának letöltése

```python
url = "https://telex.hu/"
page = requests.get(url)
soup = bs(page.text, "html.parser")

print(soup)

with open("sample_telex.html", "w", encoding="utf-8") as file:
    file.write(soup.prettify())
```

```python
f = open("sample_telex.html", "r", encoding="utf-8")
soup = bs(f, "html.parser")

print(soup)
```

## Cikkek adatait tároló kódrészlek kigyűjtése

```python
articles = soup.find_all("div", class_="item__details")

print(len(articles))
print(articles)
```

```python
print(articles[0])
```

## Egy darab cikk adatainak kinyerése

```python
category = articles[0].find("div", class_="tag__item").text.strip()
title = articles[0].find("a", class_="item__title").text.strip()
subtitle = articles[0].find("a", class_="item__lead").text.strip()
if subtitle == None:
    subtitle = ""
url = "telex.hu" + articles[0].find("a", class_="item__title").get("href")


print("Kategória: " + category)
print("Cím: " + title)
print("Alcím: " + subtitle)
print("url: " + url)
```

## Összes cikk kinyerése a főoldalról

```python
content = []

for article in articles:
    category = article.find("div", class_="tag__item").text.strip()
    title = article.find("a", class_="item__title").text.strip()
    if article.find("a", class_="item__lead") == None:
        subtitle = ""
    else:
        subtitle = article.find("a", class_="item__lead").text.strip()
    url = article.find("a", class_="item__title").get("href")

    content.append((category, title, subtitle, url))

for item in content:
    print(item)
```

## Fájlkezelés

```python
if os.path.exists("output_telex.csv"):
    os.remove("output_telex.csv")

with open("output_telex.csv", "a", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["category", "title", "subtitle", "url"])
    writer.writerows(content)
    print("Fájl írása kész!")
```
