```python
from bs4 import BeautifulSoup
import requests as re
import time
```

## Karakterek listájának letöltése

```python
url = "https://en.wikipedia.org/wiki/List_of_Ultimate_Marvel_characters"

response = re.get(url)
print(response.status_code)

if response.status_code == 200:
    print("Siker!")

soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.prettify())
```

## Linkek elkészítése

```python
# mw-content-text > div.mw-content-ltr.mw-parser-output > ul:nth-child(5) > li:nth-child(3) > a
items = soup.select(
    "#mw-content-text > div.mw-content-ltr.mw-parser-output > ul > li > a"
)

for item in items:
    print(item)
```

```python
links = []
for item in items:
    links.append("https://en.wikipedia.org" + item.get("href"))

for link in links:
    print(link)
```

## Linkek szűrése és adatgyűjtés

Amelyik linkben "#" szerepel, az egy gyűjtőlap. Azokat a karaktereket kihagyjuk.

```python
for item in range(0, 15):
    response = re.get(links[item])
    if "#" or "(" not in response:
        soup = BeautifulSoup(response.text, "html.parser")
        title = soup.select("#firstHeading > span.mw-page-title-main")[0].text
        description = (
            soup.find("div", {"class": "mw-content-ltr mw-parser-output"})
            .find_all("p")[2]
            .text
        )
        print(str(item) + ". weboldal feldolgozva:")
        print(title)
        print(description)
        open("data_marvel.csv", "a", encoding="utf-8").write(
            str(title) + "," + str(description)
        )
    else:
        time.sleep(1)
```
