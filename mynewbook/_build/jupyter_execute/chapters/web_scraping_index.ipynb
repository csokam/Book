# Index híreinek kigyűjtése

A <a href="https://index.hu/24ora/">index.hu/24ora</a> struktúrája szintén kellemes, gyakorlásra alkalmas. 

## Szükséges könyvtárak letöltése, importálása

```python
# %pip install bs4
```

```python
import os
import csv
import requests
from bs4 import BeautifulSoup as bs
```

## Weboldal HTML struktúrájának letöltése

```python
url = "https://index.hu/24ora/"
page = requests.get(url)
soup = bs(page.text, "html.parser")

print(soup)

with open("sample_index.html", "w", encoding="utf-8") as file:
    file.write(soup.prettify())
```

```python
f = open("sample_index.html", "r", encoding="utf-8")
soup = bs(f, "html.parser")

print(soup)
```

## Cikkek adatait tároló kódrészlek kigyűjtése

```python
articles = soup.find_all("article")

print(len(articles))
print(articles)
```

```python
print(articles[0])
```

## Egy darab cikk adatainak kinyerése

```python
author = articles[0].find("div", class_="szerzo").text.strip()
date = articles[0].find("div", class_="datum")
published = date.contents[0].strip()
title = articles[0].find("a").get("title")
subtitle = articles[0].find("div", class_="ajanlo").text.strip()
url = articles[0].find("a").get("href")

print("Szerző: " + author)
print("Dátum: " + published)
print("Cím: " + title)
print("Alcím: " + subtitle)
print("url: " + url)
```

## Összes cikk kinyerése a főoldalról

```python
content = []

for article in articles:
    url = article.find("a").get("href")
    if url == "http://index.hu/mindekozben":
        continue

    if article.find("div", class_="szerzo") == None:
        author = ""
    else:
        author = article.find("div", class_="szerzo").text.strip()

    if article.find("div", class_="datum") == None:
        published = ""
    else:
        date = article.find("div", class_="datum")
        published = date.contents[0].strip()

    if article.find("a").get("title") == None:
        title = ""
    else:
        title = article.find("a").get("title")

    subtitle = article.find("div", class_="ajanlo").text.strip()

    content.append((author, published, title, subtitle, url))

for item in content:
    print(item)
```

## Fájlkezelés

```python
if os.path.exists("output_index.csv"):
    os.remove("output_index.csv")

with open("output_index.csv", "a", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["author", "published", "title", "subtitle", "url"])
    writer.writerows(content)
    print("Fájl írása kész!")
```
