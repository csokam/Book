## Letöltés header fájl megadásával

```python
url = "https://www.alza.sk/lcd-monitory/18842948.htm"
header = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:109.0) Gecko/20100101 Firefox/109.0",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Charset": "utf-8",
    "Accept-Encoding": "none",
    "Referer": "http://www.google.com/",
}
page = requests.get(url, headers=header)
soup = bs(
    page.text,
    "html.parser",
)

print(soup)

with open("sample_alza_2.html", "w", encoding="utf-8") as file:
    file.write(soup.prettify())
```

## Mintafájl megnyitása

```python
f = open("sample_alza.html", "r", encoding="utf-8")
soup = bs(f, "html.parser")

print(soup)
```

## Szükséges rész meghatározása

```python
content = soup.find("div", class_="browsingitemcontainer")

print(content)
```

## Egyes termékek kigyűjtése

A HTML struktúra vizsgálata közben kiderül, hogy a használt osztály kezdete azonos, de több verzió létezik belőle. Ez a probléma megoldható többféle módszerrel, itt lambda függvény segítségével kezeljük.

```python
products = soup.find_all("div", class_=lambda x: x and x.startswith("box browsingitem"))

print(products)
```

## Egy termékre vonatkozó rész

```python
print(products[0])
```

## Egy termék adatainak kigyűjtése

```python
item = products[0].find("a", class_="name browsinglink").text
url = products[0].find("a", class_="name browsinglink").get("href")
description = products[0].find("div", class_="Description").text.strip()
price = products[0].find("span", class_="price-box__price").text

print(item)
print(url)
print(description)
print(price)
```

## Összes termék adatainak kigyűjtése

```python
items = []

for product in products:
    item = product.find("a", class_="name browsinglink").text
    url = product.find("a", class_="name browsinglink").get("href")
    description = product.find("div", class_="Description").text.strip()
    price = product.find("span", class_="price-box__price").text

    items.append((item, url, description, price))

print(items)
```

## Adatok kiírása fájlba

```python
if os.path.exists("output_alza.csv"):
    os.remove("output_alza.csv")

with open("output_alza.csv", "a", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["item", "url", "description", "price"])
    writer.writerows(items)
    print("Fájl írása kész!")

df = pd.read_csv("output_alza.csv", delimiter=",", header=0)
df.head(10)
```
