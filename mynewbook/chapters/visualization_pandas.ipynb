<h3>Simple Pandas graph</h3>

```python
import pandas as pd 

x = [1,2,3,4,5,6,7,8,9]
y = {'Temp during night': [8,12,16,15,12,9,11,13,13], 
    'Temp during day': [16,22,27,29,17,14,19,21,22]}

graph=pd.DataFrame(y,x)
graph.plot(kind="line", grid=True, title="Időjárás előrejelzés", 
        ylabel="Hőmérséklet (°C)", xlabel="Napok", xlim=(1,9), ls="dashed")
```
